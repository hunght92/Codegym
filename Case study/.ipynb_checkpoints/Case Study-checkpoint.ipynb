```python
#IMPORT CÁC THƯ VIỆN
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox


#KHAI BÁO CÁC HÀM
#Hàm lưu thông tin tài sản
def handle_luu_thong_tin():
    print("Clicked")

#Hàm sửa thông tin tài sản
def handle_sua_thong_tin():
    pass

#Hàm xác nhận xóa tài sản
def handle_xac_nhan_xoa_thong_tin():
    ket_qua = messagebox.askyesno(
        title = "Xác nhận xóa", 
        message = "Bạn có chắc chắn muốn xóa tài sản này không?"
    )
    if ket_qua:
        handle_xoa_tai_san

#Hàm xóa tài sản
def handle_xoa_tai_san():
    print("Xóa tài sản")

    
#TẠO GIAO DIỆN
#Tạo cửa sổ làm việc        
wd = tk.Tk()
wd.title("Quản lý tài sản")
wd.geometry("1920x1080")

#Tạo frame cho vùng nhập thông tin
thong_tin_frame = ttk.Frame(wd, borderwidth = 3)
thong_tin_frame.grid(row = 0, column = 0)

#Tạo vùng chọn nhóm tài sản
#Tạo nhãn
Nhom_tai_san_label = ttk.Label(
    thong_tin_frame,
    text = "Nhóm tài sản"
)

Nhom_tai_san_label.grid(
    row = 0,
    column = 0,
    pady = 3,
    sticky = "EW"
)

#Tạo Combobox để người dùng chọn nhóm tài sản
selected_group = tk.StringVar()
Nhom_tai_san_cb = ttk.Combobox(
    thong_tin_frame,
    textvariable = selected_group 
)

#Tạo list giá trị cho combobox
Nhom_tai_san_cb['value'] = [
    "Nhà cửa, công trình", 
    "Máy móc, thiết bị sản xuất", 
    "Máy móc, thiết bị khác",
    "Xe cộ",
    "Công cụ dụng cụ"    
]

#Không cho người dùng nhập thêm bộ phận khác
Nhom_tai_san_cb['state'] = 'readonly'

#Đưa combobox vào màn hình
Nhom_tai_san_cb.grid(
    row = 0,
    column = 1,
    pady = 3,
    sticky = "EW"
)

#Tạo vùng nhập mã tài sản mới
Ma_TS_label = ttk.Label(
    thong_tin_frame,
    text = "Mã tài sản"
)

Ma_TS_label.grid(
    row = 1,
    column = 0,
    pady = 3,
    sticky = "EW"
)

Ma_TS_entry = ttk.Entry(thong_tin_frame)

Ma_TS_entry.grid(
    row = 1,
    column = 1,
    pady = 3,
    sticky = "EW"
)

#Tạo vùng nhập mã tài sản cha nếu có
Ma_TS_cha_label = ttk.Label(
    thong_tin_frame,
    text = "Mã tài sản cha"
)

Ma_TS_cha_label.grid(
    row = 1,
    column = 3,
    pady = 3,
    sticky = "EW"
)

Ma_TS_cha_entry = ttk.Entry(thong_tin_frame)

Ma_TS_cha_entry.grid(
    row = 1,
    column = 4,
    pady = 3,
    sticky = "EW"
)

#Tạo vùng nhập tên tài sản mới
Ten_TS_label = ttk.Label(
    thong_tin_frame,
    text = "Tên tài sản"
)

Ten_TS_label.grid(
    row = 2,
    column = 0,
    pady = 3,
    sticky = "EW"
)

Ten_TS_entry = ttk.Entry(thong_tin_frame)

Ten_TS_entry.grid(
    row = 2,
    column = 1,
    pady = 3,
    sticky = "EW"
)

#Tạo vùng chọn bộ phận quản lý tài sản
#Tạo nhãn
Bo_phan_label = ttk.Label(
    thong_tin_frame,
    text = "Bộ phận"
)

Bo_phan_label.grid(
    row = 2,
    column = 3,
    pady = 3,
    sticky = "EW"
)

#Tạo Combobox để người dùng nhập bộ phận
selected_department = tk.StringVar()
Bo_phan_cb = ttk.Combobox(
    thong_tin_frame,
    textvariable = selected_department 
)

#Tạo list giá trị cho combobox
Bo_phan_cb['value'] = [
    "Hành chính", 
    "Kế toán", 
    "Tài chính",
    "Kế hoạch",
    "Kinh doanh", 
    "Marketing", 
    "Chuyền 1", 
    "Chuyền 2", 
    "Chuyền 3"    
]

#Không cho người dùng nhập thêm bộ phận khác
Bo_phan_cb['state'] = 'readonly'

#Đưa combobox vào màn hình
Bo_phan_cb.grid(
    row = 2,
    column = 4,
    pady = 3,
    sticky = "EW"
)

#Tạo vùng nhập ngày đưa vào sử dụng
Ngay_dua_vao_su_dung_label = ttk.Label(
    thong_tin_frame,
    text = "Ngày đưa vào sử dụng (dd/mm/yyyy)"
)

Ngay_dua_vao_su_dung_label.grid(
    row = 3,
    column = 0,
    pady = 3,
    sticky = "EW"
)

Ngay_dua_vao_su_dung_entry = ttk.Entry(thong_tin_frame)

Ngay_dua_vao_su_dung_entry.grid(
    row = 3,
    column = 1,
    pady = 3,
    sticky = "EW"
)

#Tạo vùng nhập số tháng khấu hao
So_thang_khau_hao_label = ttk.Label(
    thong_tin_frame,
    text = "Số tháng khấu hao"
)
So_thang_khau_hao_label.grid(
    row = 3,
    column = 3,
    pady = 3,
    sticky = "EW"
)

So_thang_khau_hao_entry = ttk.Entry(thong_tin_frame)

So_thang_khau_hao_entry.grid(
    row = 3,
    column = 4,
    pady = 3,
    sticky = "EW"
)

#Tạo frame cho vùng chứa các nút thao tác
thao_tac_frame = ttk.Frame(wd, borderwidth = 3)
thao_tac_frame.grid(row = 0, column = 1)

#Tạo nút lưu tài sản
save_button = ttk.Button(
    thao_tac_frame, 
    text = "Lưu", 
    command = handle_luu_thong_tin
)

save_button.grid(
    row = 0,
    column = 0,
    pady = 3,
    sticky = "EW"
)

#Tạo nút sửa thông tin
edit_button = ttk.Button(
    thao_tac_frame, 
    text = "Sửa", 
    command = handle_sua_thong_tin
)

edit_button.grid(
    row = 0,
    column = 1,
    pady = 3,
    sticky = "EW"
)

#Tạo nút xóa thông tin
delete_button = ttk.Button(
    thao_tac_frame, 
    text = "Xóa", 
    command = handle_xac_nhan_xoa_thong_tin
)

delete_button.grid(
    row = 1,
    column = 0,
    pady = 3,
    sticky = "EW"
)

#Tạo Treeview
#Khai báo tên các cột
columns = (
    "ma_tai_san_cha",
    "ma_tai_san",
    "ten_tai_san",
    "bo_phan",
    "ngay_dua_vao_su_dung",
    "so_thang_khau_hao",
    "ngay_ket_thuc_khau_hao"
)
tree = ttk.Treeview(
    wd,
    columns = columns,
    show = "headings"
)

#Khai báo tiêu đề cột
tree.heading("ma_tai_san_cha", text = "Mã tài sản cha")
tree.heading("ma_tai_san", text = "Mã tài sản")
tree.heading("ten_tai_san", text = "Tên tài sản")
tree.heading("bo_phan", text = "Bộ phận")
tree.heading("ngay_dua_vao_su_dung", text = "Ngày đưa vào sử dụng")
tree.heading("so_thang_khau_hao", text = "Số tháng khấu hao")
tree.heading("ngay_ket_thuc_khau_hao", text = "Ngày kết thúc khấu hao")

#Nhập dữ liệu
assets = []
for n in range (1,100):
    assets.append((f'ma {n}', f'ma {n}', f'ten {n}', f'bo phan {n}',f'ngay {n}', f'thang {n}', f'ngay {n}'))

#Thêm data vào treeview
for asset in assets:
    tree.insert("", tk.END, value = asset)

#Phân bổ tree
tree.grid(row = 7, column = 0, sticky = "nsew" )

#Tạo thanh cuộn
scrollbar = ttk.Scrollbar(
    wd,
    orient = tk.VERTICAL,
    command=tree.yview
)
tree.configure(yscroll=scrollbar.set)
scrollbar.grid(row=7, column=1, sticky='ns')




wd.mainloop()
```
