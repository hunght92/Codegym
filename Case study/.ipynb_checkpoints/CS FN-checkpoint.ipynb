```python
#IMPORT CÁC THƯ VIỆN
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
from datetime import date
from dateutil.relativedelta import relativedelta

#KHAI BÁO CÁC HÀM
#Hàm tính toán ngày kết thúc khấu hao
#Cần xác định thêm ngày kết thúc khấu hao dựa trên ngày bắt đầu và số tháng khấu hao
def ngay_ket_thuc_khau_hao():
    pass

#Hàm lưu thông tin tài sản
count = 0
def handle_luu_thong_tin():
#Xác định ngày kết thúc khấu hao    
    Ngay_ket_thuc_khau_hao_str = str(date(
        int(Ngay_dua_vao_su_dung_entry.get()[6:10]),
        int(Ngay_dua_vao_su_dung_entry.get()[3:5]),
        int(Ngay_dua_vao_su_dung_entry.get()[0:2])) 
        + relativedelta(months = int(So_thang_khau_hao_entry.get())))
    End_date_str = (Ngay_ket_thuc_khau_hao_str[8:10]
                    +"/"
                    +Ngay_ket_thuc_khau_hao_str[5:7]
                    +"/"
                    +Ngay_ket_thuc_khau_hao_str[0:4])
#Xử lý hàm lưu thông tin vào tree view
    global count 
    tree.insert(
        parent = "",
        index = "end",
        iid = count,
        text = "",
        values = (
            Ma_TS_cha_entry.get(),
            Ma_TS_entry.get(),
            Ten_TS_entry.get(),
            Nhom_tai_san_cb.get(),
            Bo_phan_cb.get(),
            Ngay_dua_vao_su_dung_entry.get(),
            So_thang_khau_hao_entry.get(),
            End_date_str,
            Trang_thai_cb.get()
        )
    )
    count += 1
    handle_xoa_thong_tin()

#Hàm xóa dữ liệu đang nhập
def handle_xoa_thong_tin():
    Nhom_tai_san_cb.set("")
    Ma_TS_entry.delete(0, tk.END)
    Ma_TS_cha_entry.delete(0, tk.END)
    Ten_TS_entry.delete(0, tk.END)
    Bo_phan_cb.set("")
    Ngay_dua_vao_su_dung_entry.delete(0, tk.END)
    So_thang_khau_hao_entry.delete(0, tk.END)
    Trang_thai_cb.set("")
    
#Hàm chọn tài sản
def handle_chon_tai_san():
#Xóa các ô thông tin đang có
    handle_xoa_thong_tin()
# Lấy số bản ghi trong tree view    
    tai_san_dang_chon = tree.focus()
#Lấy thông tin tài sản đang chọn (tạo ra 1 tuple)
    values = tree.item(tai_san_dang_chon,"values")    
#Cập nhật các giá trị của tuple values vào các ô thông tin
    Ma_TS_cha_entry.insert(0, values[0])
    Ma_TS_entry.insert(0, values[1])
    Ten_TS_entry.insert(0, values[2])
    Nhom_tai_san_cb.set(values[3])
    Bo_phan_cb.set(values[4])
    Ngay_dua_vao_su_dung_entry.insert(0, values[5])
    So_thang_khau_hao_entry.insert(0, values[6])
    Trang_thai_cb.set(values[8])
    
    
# Hàm cập nhật thông tin tài sản
def handle_cap_nhat_thong_tin():
#Tính lại ngày kết thúc khấu hao   
    Ngay_ket_thuc_khau_hao_str = str(date(
        int(Ngay_dua_vao_su_dung_entry.get()[6:10]),
        int(Ngay_dua_vao_su_dung_entry.get()[3:5]),
        int(Ngay_dua_vao_su_dung_entry.get()[0:2])) 
        + relativedelta(months = int(So_thang_khau_hao_entry.get())))
    End_date_str = (Ngay_ket_thuc_khau_hao_str[8:10]
                    +"/"
                    +Ngay_ket_thuc_khau_hao_str[5:7]
                    +"/"
                    +Ngay_ket_thuc_khau_hao_str[0:4])
# Lấy số bản ghi trong tree view để cập nhật thông tin  
    tai_san_dang_chon = tree.focus()
#Lưu thông tin mới
    tree.item(
        tai_san_dang_chon,
        text = "",
        values = (
            Ma_TS_cha_entry.get(),
            Ma_TS_entry.get(),
            Ten_TS_entry.get(),
            Nhom_tai_san_cb.get(),
            Bo_phan_cb.get(),
            Ngay_dua_vao_su_dung_entry.get(),
            So_thang_khau_hao_entry.get(),
            End_date_str,
            Trang_thai_cb.get()
        )
    )
#Xóa các ô thông tin đang nhập
    handle_xoa_thong_tin()
    
########################################################################################################
#TẠO GIAO DIỆN
#Tạo cửa sổ làm việc        
wd = tk.Tk()
wd.title("Quản lý tài sản")
wd.geometry("1920x1200")

#Tạo frame cho vùng nhập thông tin
thong_tin_frame = ttk.Frame(wd)
thong_tin_frame.place(x = 100, y = 50)

#Tạo vùng chọn nhóm tài sản
#Tạo nhãn
Nhom_tai_san_label = ttk.Label(
    thong_tin_frame,
    text = "Nhóm tài sản")
Nhom_tai_san_label.grid(row = 0, column = 0, pady = 3, sticky = "EW")

#Tạo Combobox để người dùng chọn nhóm tài sản
selected_group = tk.StringVar()
Nhom_tai_san_cb = ttk.Combobox(
    thong_tin_frame,
    textvariable = selected_group)

#Tạo list giá trị cho combobox
Nhom_tai_san_cb['value'] = [
    "Nhà cửa, công trình", 
    "Máy móc, thiết bị sản xuất", 
    "Máy móc, thiết bị khác",
    "Xe cộ",
    "Công cụ dụng cụ"]

#Không cho người dùng nhập thêm nhóm tài sản khác
Nhom_tai_san_cb['state'] = 'readonly'

#Đưa combobox vào màn hình
Nhom_tai_san_cb.grid(row = 0, column = 1, pady = 3, sticky = "EW")

#Tạo vùng nhập mã tài sản mới
Ma_TS_label = ttk.Label(
    thong_tin_frame,
    text = "Mã tài sản")
Ma_TS_label.grid(row = 1, column = 0, pady = 3, sticky = "EW")

Ma_TS_entry = ttk.Entry(thong_tin_frame)
Ma_TS_entry.grid(row = 1, column = 1, pady = 3, sticky = "EW")

#Tạo vùng nhập mã tài sản cha nếu có
Ma_TS_cha_label = ttk.Label(
    thong_tin_frame,
    text = "Mã tài sản cha")
Ma_TS_cha_label.grid(row = 1, column = 3, pady = 3, sticky = "EW")

Ma_TS_cha_entry = ttk.Entry(thong_tin_frame)
Ma_TS_cha_entry.grid(row = 1, column = 4, pady = 3, sticky = "EW")

#Tạo vùng nhập tên tài sản mới
Ten_TS_label = ttk.Label(
    thong_tin_frame,
    text = "Tên tài sản")
Ten_TS_label.grid(row = 2, column = 0, pady = 3, sticky = "EW")

Ten_TS_entry = ttk.Entry(thong_tin_frame)
Ten_TS_entry.grid(row = 2, column = 1, pady = 3, sticky = "EW")

#Tạo vùng chọn trạng thái tài sản
#Tạo nhãn
Trang_thai_label = ttk.Label(
    thong_tin_frame,
    text = "Trạng thái")
Trang_thai_label.grid(row = 3, column = 3, pady = 3, sticky = "EW")

#Tạo Combobox để người dùng chọn trạng thái
selected_status = tk.StringVar()
Trang_thai_cb = ttk.Combobox(
    thong_tin_frame,
    textvariable = selected_status)

#Tạo list giá trị cho combobox
Trang_thai_cb['value'] = [
    "Bình thường", 
    "Hư hỏng", 
    "Đã thanh lý"]

#Không cho người dùng nhập thêm trạng thái khác
Trang_thai_cb['state'] = 'readonly'

#Đưa combobox vào màn hình
Trang_thai_cb.grid(row = 3, column = 4, pady = 3, sticky = "EW")

#Tạo vùng chọn bộ phận quản lý tài sản
#Tạo nhãn
Bo_phan_label = ttk.Label(
    thong_tin_frame,
    text = "Bộ phận")
Bo_phan_label.grid(row = 3, column = 0, pady = 3, sticky = "EW")

#Tạo Combobox để người dùng chọn bộ phận
selected_department = tk.StringVar()
Bo_phan_cb = ttk.Combobox(
    thong_tin_frame,
    textvariable = selected_department)

#Tạo list giá trị cho combobox
Bo_phan_cb['value'] = [
    "Hành chính", 
    "Kế toán", 
    "Tài chính",
    "Kế hoạch",
    "Kinh doanh", 
    "Marketing", 
    "Chuyền 1", 
    "Chuyền 2", 
    "Chuyền 3"]

#Không cho người dùng nhập thêm bộ phận khác
Bo_phan_cb['state'] = 'readonly'

#Đưa combobox vào màn hình
Bo_phan_cb.grid(row = 3, column = 1, pady = 3, sticky = "EW")

#Tạo vùng nhập ngày đưa vào sử dụng
Ngay_dua_vao_su_dung_label = ttk.Label(
    thong_tin_frame,
    text = "Ngày đưa vào sử dụng\n(dd/mm/yyyy)")
Ngay_dua_vao_su_dung_label.grid(row = 4, column = 0, pady = 3, sticky = "EW")

Ngay_dua_vao_su_dung_entry = ttk.Entry(thong_tin_frame)
Ngay_dua_vao_su_dung_entry.grid(row = 4, column = 1, pady = 3, sticky = "EW")

#Tạo vùng nhập số tháng khấu hao
So_thang_khau_hao_label = ttk.Label(
    thong_tin_frame,
    text = "Số tháng khấu hao")
So_thang_khau_hao_label.grid(row = 4, column = 3, pady = 3, sticky = "EW")

So_thang_khau_hao_entry = ttk.Entry(thong_tin_frame)
So_thang_khau_hao_entry.grid(row = 4, column = 4, pady = 3, sticky = "EW")

#Tạo frame cho vùng chứa các nút thao tác
thao_tac_frame = ttk.Frame(wd)
thao_tac_frame.place(x = 900, y = 50)

#Tạo nút lưu tài sản
save_button = ttk.Button(
    thao_tac_frame, 
    text = "Lưu", 
    command = handle_luu_thong_tin)
save_button.pack(fill = "x")

#Tạo nút chọn tài sản
select_button = ttk.Button(
    thao_tac_frame, 
    text = "Chọn", 
    command = handle_chon_tai_san)
select_button.pack(fill = "x")

#Tạo nút cập nhật thông tin tài sản
update_button = ttk.Button(
    thao_tac_frame, 
    text = "Cập nhật", 
    command = handle_cap_nhat_thong_tin)
update_button.pack(fill = "x")

#Tạo nút xóa thông tin đang nhập
clear_button = ttk.Button(
    thao_tac_frame, 
    text = "Xóa thông tin đang nhập", 
    command = handle_xoa_thong_tin)
clear_button.pack(fill = "x")

#Tạo frame cho treeview
tree_view_frame = ttk.Frame(wd)
tree_view_frame.place(x = 50, y = 250)

#Tạo Treeview
#Khai báo tên các cột
columns = (
    "ma_tai_san_cha",
    "ma_tai_san",
    "ten_tai_san",
    "nhom_tai_san",
    "bo_phan",
    "ngay_dua_vao_su_dung",
    "so_thang_khau_hao",
    "ngay_ket_thuc_khau_hao",
    "trang_thai")

tree = ttk.Treeview(
    tree_view_frame,
    columns = columns,
    show = "headings")

#Khai báo tiêu đề cột
tree.heading("ma_tai_san_cha", text = "Mã tài sản cha")
tree.heading("ma_tai_san", text = "Mã tài sản")
tree.heading("ten_tai_san", text = "Tên tài sản")
tree.heading("nhom_tai_san", text = "Nhóm tài sản")
tree.heading("bo_phan", text = "Bộ phận")
tree.heading("ngay_dua_vao_su_dung", text = "Ngày đưa vào sử dụng")
tree.heading("so_thang_khau_hao", text = "Số tháng khấu hao")
tree.heading("ngay_ket_thuc_khau_hao", text = "Ngày kết thúc khấu hao")
tree.heading("trang_thai", text = "Trạng thái")

#Phân bổ tree
tree.grid(row = 0, column = 0, sticky = "nsew" )

#Tạo thanh cuộn
scrollbar = ttk.Scrollbar(
    tree_view_frame,
    orient = tk.VERTICAL,
    command = tree.yview)
tree.configure(yscroll=scrollbar.set)
scrollbar.grid(row = 0, column =1, sticky='ns')

wd.mainloop()
```
