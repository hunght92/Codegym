```python
import turtle
t = turtle.Turtle()


turtle.bgcolor("lightcyan")

# Draw the front side of the house
chieu_ngang_nha = 150
chieu_cao_nha = 200
t.penup()
t.goto(-700,-300)
t.pendown()
t.fillcolor("blue")
t.begin_fill()
t.fd(chieu_ngang_nha)
t.lt(90)
t.fd(chieu_cao_nha)
t.lt(90)
t.fd(chieu_ngang_nha)
t.lt(90)
t.fd(chieu_cao_nha)
t.lt(90)
t.end_fill()

# Draw the door
chieu_rong_cua = 50
chieu_cao_cua = 110
t.fd(chieu_ngang_nha/3*2)
t.fillcolor("palegreen")
t.begin_fill()
t.lt(90)
t.fd(chieu_cao_cua)
t.lt(90)
t.fd(chieu_rong_cua)
t.lt(90)
t.fd(chieu_cao_cua)
t.end_fill()

# Draw the roof
chieu_dai_mai_nha = 100
t.penup()
t.goto(-700,-100)
t.lt(131.41)
t.fillcolor("fuchsia")
t.begin_fill()
t.fd(chieu_dai_mai_nha)
t.rt(82.82)
t.fd(chieu_dai_mai_nha)
t.rt(138.59)
t.fd(chieu_ngang_nha)
t.end_fill()

# Draw the rear side of the roof
t.penup()
t.goto(-550,-100)
t.pendown()
t.fillcolor("orange")
t.begin_fill()
t.rt(41.41)
t.fd(100)
t.rt(117.18)
t.fd(120)
t.rt(62.82)
t.fd(100)
t.rt(117.18)
t.fd(120)
t.end_fill()


# Draw the side wall
t.goto(-550,-300)
t.setheading(21.41)
t.fillcolor("yellow")
t.begin_fill()
t.fd(120)
t.lt(68.59)
t.fd(200)
t.lt(111.41)
t.fd(120)
t.end_fill()

# Draw the windows
t.penup()
t.goto(-500,-200)
t.pendown()
t.setheading(21.41)
t.fillcolor("peru")
t.begin_fill()
t.fd(30)
t.lt(68.59)
t.fd(50)
t.lt(111.41)
t.fd(30)
t.lt(68.59)
t.fd(50)
t.end_fill()

# Draw the tree
t.penup()
t.goto(0,-300)
t.setheading(0)
t.fillcolor("brown")
t.begin_fill()
t.fd(40)
t.lt(90)
t.fd(90)
t.lt(90)
t.fd(40)
t.lt(90)
t.fd(90)
t.end_fill()

# Vẽ tán cây 1
t.penup()
t.goto(-40,-210)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.setheading(39.81)
t.fd(78.1)
t.rt(79.62)
t.fd(78.1)
t.end_fill()

# Vẽ tán cây 2
t.penup()
t.goto(-30,-160)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.setheading(38.66)
t.fd(64.03)
t.rt(77.32)
t.fd(64.03)
t.end_fill()

# Vẽ tán cây 3
t.penup()
t.goto(-20,-120)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.setheading(36.87)
t.fd(50)
t.rt(73.74)
t.fd(50)
t.end_fill()

# Draw the sun
t.penup()
t.goto(-250,250)
t.pendown()
t.setheading(0)
t.fillcolor("yellow")
t.begin_fill()
t.circle(35)
t.end_fill()

t.penup()
t.goto(-215,285)
t.pendown()
t.fd(50)

t.penup()
t.goto(-225.25,309.75)
t.pendown()
t.setheading(45)
t.fd(30)

t.penup()
t.goto(-250,320)
t.pendown()
t.setheading(90)
t.fd(50)

t.penup()
t.goto(-274.75,309.75)
t.pendown()
t.setheading(135)
t.fd(30)

t.penup()
t.goto(-285,285)
t.pendown()
t.setheading(180)
t.fd(50)

t.penup()
t.goto(-274.75,260.25)
t.pendown()
t.setheading(225)
t.fd(30)

t.penup()
t.goto(-250,250)
t.pendown()
t.setheading(270)
t.fd(50)

t.penup()
t.goto(-225.25,260.25)
t.pendown()
t.setheading(315)
t.fd(30)


turtle.done()
```
