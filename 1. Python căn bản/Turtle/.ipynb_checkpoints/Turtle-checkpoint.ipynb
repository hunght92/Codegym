```python
import turtle
h = turtle.Turtle()

canh = 100
khoang_cach = 40

h.speed(2)
#move con tro ra vi tri moi de hinh ve doi xung o tam
h.penup()
h.forward(canh/2)
h.left(90)
h.pendown()

# ve hinh vuong 1, canh 100, mau do
h.pencolor('red')
h.forward(canh/2)
h.left(90)
h.forward(canh)
h.left(90)
h.forward(canh)
h.left(90)
h.forward(canh)
h.left(90)
h.forward(canh/2)

# move con tro ra vi tri moi cach canh hinh vuong 1 40px
h.penup()
h.right(90)
h.forward(khoang_cach)
h.left(90)
h.pendown()

# ve hinh vuong 2, canh 100 + 40*2 = 180, mau vang
h.pencolor('yellow')
h.forward((canh+khoang_cach*2)/2)
h.left(90)
h.forward(canh+khoang_cach*2)
h.left(90)
h.forward(canh+khoang_cach*2)
h.left(90)
h.forward(canh+khoang_cach*2)
h.left(90)
h.forward((canh+khoang_cach*2)/2)

# move con tro ra vi tri moi cach canh hinh vuong 2 40px
h.penup()
h.right(90)
h.forward(khoang_cach)
h.left(90)
h.pendown()

#
h.forward(130)
h.left(90)
h.forward(260)
h.left(90)
h.forward(260)
h.left(90)
h.forward(260)
h.left(90)
h.forward(130)



turtle.done()
```
