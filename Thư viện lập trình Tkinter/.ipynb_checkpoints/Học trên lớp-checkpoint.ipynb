```python
import tkinter as tk
from tkinter import ttk

main = tk.Tk()
lb = tk.Listbox(main)
lb.insert(tk.END, "Item 1") #0
lb.insert(tk.END, "Item 2") #1
lb.insert(tk.END, "Item 3") #2
# lb.delete(1,2)

lb.grid(row = 0,column = 0)

def handle_click_btn():
    print(lb.curselection())

btn = ttk.Button(main, text = "Click", command = handle_click_btn)
btn.grid(row = 1, column = 0)   

main.mainloop()
```
