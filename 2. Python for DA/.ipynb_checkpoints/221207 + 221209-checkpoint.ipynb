```python
# Từ khóa LIMIT: giới hạn dữ liệu trả về
# Cú pháp:
#......
# LIMIT số lượng trả về
```

```python
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' OR country = 'USA' \
ORDER BY city ASC \
LIMIT 3"
vidu9 = pd.read_sql_query(query, connection)
vidu9
```

```python
# Truy vấn tên khách hàng, nước và thành phố và creditLimit từ bảng customers mà nước là UK, có creditLimit là lớn nhất
query = "SELECT customerName, country, city, creditLimit FROM customers \
WHERE country = 'UK' \
ORDER BY creditLimit DESC \
LIMIT 1"

vidu10 = pd.read_sql_query(query, connection)
vidu10
```

```python
# DISTINCT
# Có tác dụng tương đương với unique() trong pandas, trả về các giá trị đơn nhất
# Truy xuất tên các nước từ bảng customers
query = "SELECT DISTINCT country FROM customers"

vidu11 = pd.read_sql_query(query, connection)
vidu11
```

```python
# Truy xuất tên các thành phố của Nga từ bảng customers
query = "SELECT DISTINCT city FROM customers WHERE country = 'UK'"

vidu12 = pd.read_sql_query(query, connection)
vidu12
```

```python
# AGGREGATE FUNCTIONS
# Hàm COUNT để đếm số lượng
# Truy xuất số lượng các thành phố của Japan từ bảng customers
query = "SELECT COUNT(DISTINCT city) FROM customers WHERE country = 'Germany'"

vidu13 = pd.read_sql_query(query, connection)
vidu13
```

```python
query = "SELECT COUNT(DISTINCT customerNumber) FROM customers WHERE country = 'USA'"

vidu14 = pd.read_sql_query(query, connection)
vidu14
```

```python
# HÀM SUM
# Dùng để tính tổng
# Tính tổng creditLimit của các khách hàng đến từ USA
query = "SELECT SUM(creditLimit) FROM customers WHERE country = 'USA'"

vidu15 = pd.read_sql_query(query, connection)
vidu15
```

```python
# HÀM MIN
# Tính giá trị nhỏ nhất
# Tính creditLimit nhỏ nhất của các khách hàng đến từ USA
query = "SELECT MIN(creditLimit) FROM customers WHERE country = 'USA'"

vidu16 = pd.read_sql_query(query, connection)
vidu16
```

```python
# HÀM MIN
# Tính giá trị lớn nhất
# Tính creditLimit lớn nhất của các khách hàng đến từ USA
query = "SELECT MAX(creditLimit) FROM customers WHERE country = 'USA'"

vidu17 = pd.read_sql_query(query, connection)
vidu17
```

```python
# HÀM AVERAGE
# Tính giá trị trung bình
# Tính creditLimit trung bình của các khách hàng đến từ USA
query = "SELECT AVG(creditLimit) FROM customers WHERE country = 'USA'"

vidu18 = pd.read_sql_query(query, connection)
vidu18
```
