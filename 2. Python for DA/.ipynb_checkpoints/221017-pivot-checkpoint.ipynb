```python
# Th√™m c·ªôt "rating", v·ªõi c√¥ng th·ª©c l√† rating = rating_good*2 + rating_normal - rating_bad*3
data.loc[:,'rating'] = data.loc[:, 'rating_good'] * 2 + data.loc[:, 'rating_normal'] - data.loc[:, 'rating_bad'] * 3
data
```

```python
# Th√™m c·ªôt 'location' c√≥ c·∫•u tr√∫c nh∆∞ sau: 'shop_location, Philippine'
data.loc[:, 'location'] = data.loc[:, 'shop_location'] + ', Philippine'
data
```

```python
# Th√™m c·ªôt 'join' c√≥ c·∫•u tr√∫c 'Month day, year'
# G·ª£i √Ω: D√πng astype(str) ƒë·ªÉ chuy·ªÉn s·ªë th√†nh string
data.loc[:, 'join'] = data.loc[:, 'join_month'].astype(str) + ' ' + data.loc[:, 'join_day'].astype(str) + ', ' + data.loc[:, 'join_year'].astype(str)
data
```

```python
# Th√™m c·ªôt 'test1' v·ªõi gi√° tr·ªã logic l√† shop m·ªü sau 2020
data.loc[:, 'test1'] = data.loc[:, 'join_year'] > 2020
data
```

```python
# Th√™m c·ªôt 'medal' v·ªõi gi√° tr·ªã 'gold' n·∫øu rating > 10000 v√† 'silver' n·∫øu <= 10000
data.loc[:, 'medal'] = np.where(data.loc[:, 'rating'] > 10000, 'gold', 'silver')
data
```

```python
# S·ª≠a c·ªôt 'medal' v·ªõi gi√° tr·ªã 'gold' n·∫øu rating >= 100000, 'silver' n·∫øu 50000 <= rating < 100000 v√† 'bronze' n·∫øu 10000 <= rating < 50000 v√† 'iron' n·∫øu rating < 10000
conditions = [(data.loc[:, 'rating'] >= 100000),
              (data.loc[:, 'rating'] >= 50000) & (data.loc[:, 'rating'] < 100000),
              (data.loc[:, 'rating'] >= 10000) & (data.loc[:, 'rating'] < 50000)]
values = ['gold', 'silver', 'bronze']
data.loc[:, 'medal'] = np.select(conditions, values, default = 'iron')
data
```
