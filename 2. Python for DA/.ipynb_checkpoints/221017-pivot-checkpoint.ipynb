```python
# Thêm cột "rating", với công thức là rating = rating_good*2 + rating_normal - rating_bad*3
data.loc[:,'rating'] = data.loc[:, 'rating_good'] * 2 + data.loc[:, 'rating_normal'] - data.loc[:, 'rating_bad'] * 3
data
```

```python
# Thêm cột 'location' có cấu trúc như sau: 'shop_location, Philippine'
data.loc[:, 'location'] = data.loc[:, 'shop_location'] + ', Philippine'
data
```

```python
# Thêm cột 'join' có cấu trúc 'Month day, year'
# Gợi ý: Dùng astype(str) để chuyển số thành string
data.loc[:, 'join'] = data.loc[:, 'join_month'].astype(str) + ' ' + data.loc[:, 'join_day'].astype(str) + ', ' + data.loc[:, 'join_year'].astype(str)
data
```

```python
# Thêm cột 'test1' với giá trị logic là shop mở sau 2020
data.loc[:, 'test1'] = data.loc[:, 'join_year'] > 2020
data
```

```python
# Thêm cột 'medal' với giá trị 'gold' nếu rating > 10000 và 'silver' nếu <= 10000
data.loc[:, 'medal'] = np.where(data.loc[:, 'rating'] > 10000, 'gold', 'silver')
data
```

```python
# Sửa cột 'medal' với giá trị 'gold' nếu rating >= 100000, 'silver' nếu 50000 <= rating < 100000 và 'bronze' nếu 10000 <= rating < 50000 và 'iron' nếu rating < 10000
conditions = [(data.loc[:, 'rating'] >= 100000),
              (data.loc[:, 'rating'] >= 50000) & (data.loc[:, 'rating'] < 100000),
              (data.loc[:, 'rating'] >= 10000) & (data.loc[:, 'rating'] < 50000)]
values = ['gold', 'silver', 'bronze']
data.loc[:, 'medal'] = np.select(conditions, values, default = 'iron')
data
```
