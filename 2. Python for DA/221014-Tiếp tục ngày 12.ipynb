```python
# Hàm concat
# pandas.concat([data1, data2,...], axis, join)
# data1, data2... là các bảng cần gộp
# axis = 1 là ghép dọc (thêm cột) còn axis = 0 nghĩa là ghép trên dưới (thêm hàng)
# join = "outer" nghĩa là lấy toàn bộ và nếu trùng lặp thì chỉ giữ 1, join = "inner" nghĩa là chỉ lấy những cái lặp
```

```python
data
```

```python
data1 = data.loc[:, :"UmId"]
data2 = data.loc[:, "UmName":]
display(data1, data2)
```

```python
data121 = pd.concat([data1, data2], axis = 0, join = "outer")
display(data121)
```

```python
data122 = pd.concat([data1, data2], axis = 0, join = "inner")
display(data122)
```

```python
data123 = pd.concat([data1, data2], axis = 1, join = "outer")
display(data123)
```

```python
data124 = pd.concat([data1, data2], axis = 1, join = "inner")
display(data124)
```

```python
# Ghép theo cột - Hàm merge() - Tương đương hàm vlookup trong excel
# data1.merge(data2, on = "cột chung")
# data1, data2 là 2 bảng cần gộp. Chỉ merge được 2 bảng "cột chung" là các cột sẽ được giữ 1 bản duy nhất
```

```python
data1 = data.loc[:, :"ProductName"]
data2 = data.loc[:, "ProductName":]
display(data1, data2)
```

```python
data125 = data1.merge(data2, on = "ProductName")
data125
```
