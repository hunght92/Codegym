```python
# Xử lý giá trị khuyết thiếu ở cột CustomerID: Xóa các dòng chứa NaN tại cột description, thay thế giá trị khuyết thiếu của CustomerID bằng giá trị "GUEST"
data = data.dropna(subset = 'Description')
data
```

```python
# Thay các giá trị khuyết thiếu của cột CustomerID bằng giá trị 'GUEST'
values = {'CustomerID':'GUEST'}
data = data.fillna(value = values)
data.info()
```

```python
# Thực hiện tìm giá trị ngoại lai của thuộc tính Quantity và thuộc tính Unit Price
import seaborn as sns
import matplotlib.pyplot as plt

# sns.histplot(data2['Quantity'])
plt.hist(data['Quantity'], bins = 100)
plt.show()
```

```python
# Dữ liệu xoay xung quanh 0, nhưng có những giá trị rất lớn hoặc rất nhỏ
# Bước đầu tiên là loại các giá trị âm
data4 = data.loc[data.loc[:, 'Quantity'] > 0, :]
data4
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# sns.histplot(data2['Quantity'])
plt.hist(data4['Quantity'], bins = 100)
plt.show()
```

```python
# Sử dụng phương pháp tứ phân vị IQR để xác định các giá trị ngoại lai của cột Quantity
Q1_quantity = data4.loc[:, 'Quantity'].quantile(0.25, 'linear')
Q3_quantity = data4.loc[:, 'Quantity'].quantile(0.75)
IQR_quantity = Q3_quantity - Q1_quantity
upper_quantity = (Q3_quantity + 2 * IQR_quantity)
lower_quantity = (Q1_quantity - 2 * IQR_quantity)
print(f'Upper bound quantity: {upper_quantity}')
print(f'Lower bound quantity: {lower_quantity}')

sns.boxplot(x = data4['Quantity'])
# print(data4.loc[:, 'Quantity']\
#      [(data4['Quantity'] > upper_quantity) | (data4['Quantity'] < lower_quantity)])
print(data4.loc[:, 'Quantity']\
     [(data4['Quantity'] > upper_quantity) | (data4['Quantity'] < lower_quantity)].count())
```

```python
data5 = data4.loc[:, :][(data4['Quantity'] < upper_quantity) & (data4['Quantity'] > lower_quantity)]
data5
```

```python
plt.hist(data5['Quantity'], bins = 10)
plt.show()
```

```python
plt.hist(data['UnitPrice'], bins = 10)
plt.show()
```

```python
data6 = data5.loc[data5.loc[:, 'UnitPrice'] > 0, :]
data6
```

```python
# Sử dụng phương pháp tứ phân vị IQR để xác định các giá trị ngoại lai của cột UnitPrice
Q1_price = data6.loc[:, 'UnitPrice'].quantile(0.25, 'linear')
Q3_price = data6.loc[:, 'UnitPrice'].quantile(0.75)
IQR_price = Q3_price - Q1_price
upper_price = (Q3_price + 1.5 * IQR_price)
lower_price = (Q1_price - 1.5 * IQR_price)
print(f'Upper bound unit price: {upper_price}')
print(f'Lower bound unit price: {lower_price}')

sns.boxplot(x = data6['UnitPrice'])
# print(data4.loc[:, 'Quantity']\
#      [(data4['Quantity'] > upper_quantity) | (data4['Quantity'] < lower_quantity)])
print(data6.loc[:, 'UnitPrice']\
     [(data6['UnitPrice'] > upper_price) | (data6['UnitPrice'] < lower_price)].count())
```

```python
data7 = data6.loc[:, :][(data6['UnitPrice'] < upper_price) & (data6['UnitPrice'] > lower_price)]
plt.hist(data7['UnitPrice'], bins = 10)
plt.show()
```
