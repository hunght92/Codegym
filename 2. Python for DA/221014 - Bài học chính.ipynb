```python
# L·ªçc l·∫•y d·ªØ li·ªáu c·ªßa c√°c shop c√≥ tr√™n 5 nƒÉm t·ªìn t·∫°i, ƒë∆∞·ª£c shopee verify v√† kh√¥ng c√≥ qu√° 200 ƒë√°nh gi√° x·∫•u
data[(data["is_shopee_verified"] == 1) 
     & (data["rating_bad"] < 200) 
     & (2023 - (data["join_year"]) > 5)]
```

```python
# L·ªçc b·∫±ng h√†m query
# Ph∆∞∆°ng th·ª©c truy v·∫•n query gi√∫p c√¢u l·ªánh truy v·∫•n ch·∫°y nhanh h∆°n
# C√°c ƒëi·ªÅu ki·ªán ƒë∆∞·ª£c n·ªëi v·ªõi nhau b·∫±ng c√°c ph√©p to√°n logic nh∆∞ and (v√†), or (ho·∫∑c), not (ph·ªß ƒë·ªãnh) v√† ƒë∆∞·ª£c ƒë·ªÉ trong m·ªôt c·∫∑p d·∫•u nh√°y k√©p
# df.query("(T√™n c·ªôt 1 PTSS Gi√° tr·ªã 1) PTLG (T√™n c·ªôt n PTSS Gi√° tr·ªã n)")
# df.query("(So s√°nh 1) PTLG (So s√°nh 2)")
```

```python
# L·ªçc l·∫•y t√™n c√°c shop v√† ƒë·ªãa ch·ªâ c√≥ tr√™n 100 ƒë√°nh gi√° t·ªët v√† trung b√¨nh ƒë√°nh gi√° tr√™n 4.8 sao
data.query("(rating_good > 1000) and (rating_star > 4.8)")\
.loc[:,["name", "shop_location"]]
```

```python
# L·ªçc l·∫•y d·ªØ li·ªáu c√°c shop c√≥ t·ª´ 5 nƒÉm t·ªìn t·∫°i, ƒë∆∞·ª£c verify v√† kh√¥ng c√≥ qu√° 200 ƒë√°nh gi√° x·∫•u
data.query("((2023 - join_year) > 5) and (is_shopee_verified == 1) and (rating_bad <200)") # c√≥ th·ªÉ kh√¥ng c√≥ () v·∫´n ho·∫°t ƒë·ªông ƒë∆∞·ª£c
```
