```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("FoodPrice_in_Turkey.csv", header = 0, encoding = "ISO-8859-1")
display(data)
```

```python
data.info()
```

```python
# Thêm 2 dòng có index -1, 2.5 vào bảng
data.loc[-1] = ['ThanhHoa', 66, 'Nem chua', 5, 'KG', 6, 2022, 66.1234]
data.loc[2.5] = ['DaNang', 77, 'BTCTH', 5, 'KG', 6, 2022, 77.1234]
data = data.sort_index()
display(data.head(10))
```

```python
# Truy xuất 1 dòng
# Ví dụ: Lấy dòng có chứa Đà Nẵng
data.loc[2.5]
```

```python
# Lấy dòng có chứa Thanh Hóa
data.loc[-1]
```

```python
# Lấy dòng có chứa Đà Nẵng dùng iloc, Chú ý hàng đầu tiên đếm từ 0
# Nếu biết index thì dùng loc nhanh hơn, còn iloc thì dùng lúc luôn luôn cần lấy dòng đầu tiên sẽ tiện hơn
data.iloc[4]
```

```python
# Lấy dòng có chứa Thanh Hóa dùng iloc
data.iloc[0]
```

```python
# Lấy dòng chứa Thanh Hóa và Đà Nẵng dùng loc
data.loc[[-1, 2.5]]
```

```python
# Lấy dòng chứa Thanh Hóa và ĐN dùng iloc
data.iloc[[0, 4]]
```

```python
# Truy xuất nhiều dòng liền nhau từ Thanh Hóa đến Đà Nẵng dùng loc
data.loc[-1:2.5]
```

```python
# Truy xuất nhiều dòng liền nhau từ Thanh Hóa đến Đà Nẵng dùng iloc
data.iloc[0:5]
```

```python
# Nếu thiếu vị trí bắt đầu thì mặc định là bắt đầu từ đầu bảng (0)
# Nếu thiếu vị trí kết thúc thì mặc định là kết thúc ở cuối bảng
# Tương tự với loc
data.iloc[:5]
```

```python
data.loc[2.5:]
```

```python
# Index bằng chữ thì chỉ sử dụng được loc
# Để dùng iloc thì index luôn luôn phải là số (i = integer)
```

```python
# Truy xuất tới 1 cột, hệ thống sẽ trả về 1 serie
data.loc[-1:2.5, "ProductName"]
```

```python
# Chọn cột có tên place
data.loc[:, "Place"]
```

```python
# Chọn cột dùng iloc, index luôn bắt đầu từ 0
data.iloc[:, 2]
```

```python
data.iloc[:, 7]
```

```python
# Truy xuất tới nhiều cột rời rạc
data.loc[:, ["ProductName", "Place"]]
```

```python
data.iloc[[0,5], [2, 0]]
```

```python
# Truy xuất tới nhiều cột liên tiếp
data.loc[:,"ProductId":"UmName"]
```

```python
# Nếu dùng iloc thì phải lưu ý số cuối + 1
data.iloc[:, 1:5]
```

```python
# Lấy một số ô nhất định
# 
data.loc[2:5,"ProductId":"Year"]
```

```python
data.loc[9] = ['NhaTrang', 88, 'Cá ngựa khô', 5, 'KG', 10, 2022, 77.1234]
data = data.sort_index()
data.head(15)
```

```python
# Lấy cột có chứa từ Place đến ProductName, 10 hàng bắt đầu từ Đà Nẵng
data.loc[2.5:11,"Place":"ProductName"]
```

```python
# Lấy cột có chứa từ Place đến ProductName, 10 hàng bắt đầu từ Đà Nẵng
data.iloc[4:14,0:3]
```

```python
data.loc[9,"ProductName"]
```

```python
# Bonus: Slicer
data.iloc[0:5:2, 0:7:2]
# Cấu trúc selector a:b:c => a: bắt đầu, b: kết thúc + 1 (hoặc bắt đầu + số lượng), c: Khoảng cách
```

```python
# Sử dụng chained indexing (chỉ nên dùng để truy vấn)
# Truy xuất giá trị: df.['tên cột'][index] với index là chỉ số dòng
# Tuy nhiên khi set giá trị thì hệ thống sẽ lưu lên vùng nhớ mới, nên không nên dùng để set dữ liệu
```

```python
# Sử dụng at
data.at[0,"Place"]
```

```python
# Sử dụng at cũng gán được dữ liệu
data.at[0,"Place"] = "SaiGon"
data
```
