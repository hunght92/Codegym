```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("FoodPrice_in_Turkey.csv", header = 0, encoding = "ISO-8859-1")
data
```

```python
data.columns
```

```python
# Việt hóa tên các cột bằng cách đổi tên
# Chỉ đổi tên 1 cột
data.rename(columns = {"Place":"Địa điểm"}, inplace = True) #inplace sẽ thay thế trực tiếp vào data gốc. Bản gốc dataframe được lưu trên bộ nhớ RAM
data
```

```python
# Đổi tên vài cột xác định
data.rename(columns = {"Place":"Địa điểm", "ProductName":"Tên sản phẩm"}, inplace = True)
data
```

```python
# Đổi tên toàn bộ cột (Cần liệt kê đầy đủ tên cột mới và theo thứ tự)
data.columns = ["Địa điểm", "Mã sản phẩm", "Tên sản phẩm", "Đơn vị", "Tên đơn vị", "Tháng", "Năm", "Giá"]
data
```

```python
# Thêm cột vào bảng đã có
```

```python
# Mặc định thêm cột mới thì cột mới sẽ được thêm vào cuối bảng
data["Tên cột mới"] = np.NaN
data
```

```python
# Chèn thêm cột "Chiết khấu" vào sau cột "Giá"
data.insert(8, "Chiết khấu (%)", 20)
data
```

```python
# Thêm cột số lượng trước cột "Giá", giá trị đều bằng 50
data.insert(7, "Số lượng", 50)
data
```

```python
# Đổi tên cột mới thành "Tổng" và có giá trị bằng Giá x Số lượng x (100% - Chiết khấu)
data.rename(columns = {"Tên cột mới":"Tổng"}, inplace = True)
data
```

```python
data["Tổng"] = data["Giá"] * data["Số lượng"] * (100 - data["Chiết khấu (%)"])/100
data
```

```python
# Thêm 1 dòng mới vào cuối bảng
data.shape[0] #lấy được chỉ số dòng cuối
# data.shape[1] #lấy được chỉ số cột cuối
```

```python
data.loc[data.shape[0]] = ["Sài Gòn", 3000, "Dưa hấu", 5, "KG", 10, 2022, 50, 50.000, 10, np.NaN]
data
# data.info()
```

```python
# Thêm dòng mới vào đầu bảng
data.loc[-1] = ["Hà Nội", 3001, "Cốm làng Vòng", 5, "KG", 10, 2022, 20, 50.000, 30, np.NaN]
# Sắp xếp lại bảng
data = data.sort_index()
# Update lại index của bảng
data.index = data.index + 1
data
```

```python
# Thêm một dòng mới ở giữa bảng: Sau dòng HN, với các giá trị lần lượt là 
data.loc[0.5] = ["Đà Nẵng", 3002, "Mỳ quảng", 5, "Tô", 10, 2022, 20, 100.000, 10, np.NaN]
data = data.sort_index().reset_index(drop = True)
data
```
